```python
def symbol(s,n,t,n_l=''):
    return '\t'*t+s*n+n_l

def string(s,t,n_l=''):
    return '\t'*t+s+n_l

def Dec2Bin(x):
    result = []
    if x>0:
        while(x>0):
            rem = x % 2
            x //= 2
            result.append(rem)

        if len(result)==3:
            result.insert(4,0)

        elif len(result)==2:

            for i in range(2,0,-1):
                result.insert(i+1,0)

        elif len(result)==1:

            for i in range(3,0,-1):
                result.insert(i+1,0)
    elif x==0:
        result = [0,0,0,0]
    
    else:
        print('\n\t\t\tNO BINARY FOR NEGITIVE NUMBERS')
        
    return result

def Bin2Dec(x):
    result = 0
    
    for i in range(len(str(x))):
        digit = x % 10
        
        if digit==1:
            result += digit*pow(2,i)
    
        x //= 10
    
    return result

def Dec2Oct(x):
    result = []
    if x>0:
        while(x>0):
            rem = x % 8
            x //= 8
            result.append(rem)
    elif x==0:
        result = [0]
    
    else:
        print('\n\t\tNO OCTAL FOR NEGITIVE NUMBERS')
        
    return result

def Oct2Dec(x):
    result = 0
    for i in range(len(str(x))):
        digit = x % 10
        result += digit*pow(8,i)
        x //= 10
    return result

myTable = PrettyTable()
myTable.field_names = ["OPTIONS", "CONVERSIONS"]
print(symbol('-',35,3,'\n')+string('WHAT TYPE OF CONVERSION DO YOU WANT',3,'\n')+symbol('-',35,3,'\n'))

count = 1
for i in 'DECIMAL TO BINARY','BINARY TO DECIMAL','DECIMAL TO OCTAL','OCTAL TO DECIMAL':
    myTable.add_row([count,i])
    count += 1
print(myTable)

choice=0

while(choice==0 or choice=='1' or choice=='2' or choice=='3' or choice=='4'):
    choice = input('\nEnter One Of The Above Conversion Choices or Enter any key to exit => ')

    if choice == '1':
        print()
        print(symbol('-',50,3))
        number = int(input("\t\t\tENTER THE NUMBER IN DECIMAL FORM => "))

        result = Dec2Bin(number)

        print('\n\t\t\tTHE BINARY FORM OF (',number,') IS =>  ',end=' ')

        for i in result[::-1]:
            print(i,end=' ')

        print()
        print(symbol('-',50,3))

    elif choice == '2':
        print()
        print(symbol('-',50,3))
        number = int(input("\t\t\tENTER THE NUMBER IN BINARY FORM => "))

        result = Bin2Dec(number)

        print('\n\t\t\tTHE DECIMAL FORM OF (',number,') IS =>  ',result)
        print(symbol('-',50,3))

    elif choice == '3':
        print()
        print(symbol('-',50,3))
        number=int(input("\t\t\tENTER THE NUMBER IN DECIMAL FORM => "))

        result = Dec2Oct(number)

        print('\n\t\t\tTHE OCTAL FORM OF (',number,') IS =>  ',end=' ')

        for i in result[::-1]:
            print(i,end=' ')

        print()
        print(symbol('-',50,3))

    elif choice == '4':
        print()
        print(symbol('-',50,3))
        number = int(input("\t\t\tENTER THE NUMBER IN OCTAL FORM => "))

        result = Oct2Dec(number)

        print('\n\t\t\tTHE DECIMAL FORM OF (',number,') IS =>  ',result)
        print(symbol('-',50,3))
print()
print(symbol('-',26,3,'\n')+string('THANK YOU -------> THE END',3,'\n')+symbol('-',26,3,'\n'))
```
