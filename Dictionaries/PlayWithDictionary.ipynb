```python
# Please Install and import these functions, They are just for printing a decent table
# Install the following package instllation,(No need in jupyter ,just run as it is)
#!pip install prettytable
from prettytable import PrettyTable

# Your code for Add_Student goes here
def Add_Student(d1,d2):
    d3 = {}
    # Adding two dictionaries and storing the data in third one
    for i in d1:
        d3[i] = d1[i]
    for i in d2:
        d3[i] = d2[i]
    
    newTable = PrettyTable()
    newTable.field_names = ["ROLL NUMBER","FULL NAME","ADMISSION","CGPA","SEMESTER"]
    # printing
    for i in d3:
        newTable.add_row([i,d3[i][0] + ' ' + d3[i][1],d3[i][2],d3[i][3],d3[i][4]])
    
    print(newTable)
    return d3

# Your code for Search_Student goes here
def Search_Student(d,n):
    d4={}
    n[0]=n[0].capitalize()
    n[1]=n[1].capitalize()
    
    for i in d:
        if (n[0] == d[i][0]) and (n[1] == d[i][1]):
            d4[i]=d[i]
            n=' '.join(n)
            print('\n',n,"is in the dictionary => ",d4)
            break
    else:
        print('\nEntered Student is not in the dictionary => ',d4)
    
    return d4

# Your code for Update_Student goes here
def Update_Student(s,u):
    updatedTable = PrettyTable()
    updatedTable.field_names = ["ROLL NUMBER","FULL NAME","ADMISSION","CGPA","SEMESTER"]
    for i in s:
        if i==u[0]:
            if u[1]=='first name':
                s[i][0]=u[2].capitalize()
            
            elif u[1]=='last name':
                s[i][1]=u[2].capitalize()
            
            elif u[1]=='status':
                u[2]=u[2].capitalize()
                
                if u[2]=='True':
                    u[2]=True
                else:
                    u[2]=False
                
                s[i][2]=u[2]
            
            elif u[1]=='cgpa':
                s[i][3]=float(u[2])
            
            elif u[1]=='semester':
                s[i][4]=int(u[2])
        
        updatedTable.add_row([i,s[i][0] + ' ' + s[i][1],s[i][2],s[i][3],s[i][4]])
    
    print(updatedTable)
    return s


# Your code for Delete_Student goes here
def Delete_Student(s,dell):
    deleteTable = PrettyTable()
    deleteTable.field_names = ["ROLL NUMBER","FULL NAME","ADMISSION","CGPA","SEMESTER"]
    dell[0]=dell[0].capitalize()
    dell[1]=dell[1].capitalize()
    
    for i in s:
        if (s[i][0]==dell[0]) and (s[i][1]==dell[1]):
            s.pop(i)
            break
    for i in s:
        deleteTable.add_row([i,s[i][0] + ' ' + s[i][1],s[i][2],s[i][3],s[i][4]])
    
    print(deleteTable)
    return s

# Your code for Get_Students goes here
def Get_Students(s):
    same_gpa = []; same_semester = []; same_gpa_or_semester = []; get_student  = []
    
    for i in s:
        for j in s:
            if s[j][3] == s[i][3] and s[j][0] != s[i][0]:
                if s[i][0]+' '+s[i][1] not in same_gpa:
                    same_gpa.append(s[i][0]+' '+s[i][1])
            
            if s[j][4] == s[i][4] and s[j][0] != s[i][0]:
                if s[i][0]+' '+s[i][1] not in same_semester:
                    same_semester.append(s[i][0]+' '+s[i][1])
            
            if (s[j][3] == s[i][3] or s[j][4] == s[i][4]) and s[j][0] != s[i][0]:
                if s[i][0]+' '+s[i][1] not in same_gpa_or_semester:
                    same_gpa_or_semester.append(s[i][0]+' '+s[i][1])
    
    for i in same_gpa,same_semester,same_gpa_or_semester:
        get_student.append(i)
    
    print(get_student)
    return get_student

# Rest of the code (Global scope)
student = { 
    "p21-8030" : ["Muawiya","Kaleem",True,3.5,1],
    "p21-8031" : ["Shehryar","Sher",False,3.3,2],
    "p21-8032" : ["Ali","Sher",True,3,3],
    "p21-8033" : ["Usama","Jameel",True,2.8,4],
    "p21-8034" : ["Fahad","Jalal",False,3.4,5],
    "p21-8035" : ["Tayyaba","Bibi",True,3.1,6],
    "p21-8036" : ["Afshan","Naeem",False,2.6,2],
    "p21-8037" : ["Kashif","Khan",True,2.4,1],
    "p21-8038" : ["Rohma","Khan",True,3.1,4],
    "p21-8039" : ["Maryam","Nawaz",True,3.5,7],
}

new_students = {}
count = 1
p = 'Y'
print('\n\t\t\tENTER RECORDS OF NEW STUDENTS')
while (p == 'Y'):
    values = []
    print('\n\t\tENTER DETAIL ABOUT STUDENT {a}:\n'.format(a=count))
    
    roll_number = input("Enter the Roll Number => ")
    first_name = input("Enter the First Name => ").capitalize()
    last_name = input("Enter the Last Name => ").capitalize()
    status = input("Enter the Status of Admission(True or False) => ").capitalize()
    cgpa = float(input("Enter the CGPA => "))
    semester = int(input("Enter the Semester Number => "))
    
    if status=='True':
        status = True
    elif status=='False':
        status = False
    
    for i in first_name,last_name,status,cgpa,semester:
        values.append(i)
    
    new_students[roll_number] = values
    count+=1
    
    #call for the ADD Function
    choice_1 = input('\nTo End the student Record Entry, Enter n/N\nOR\nTo ADD more Students Enter any key => ').upper()
    
    if choice_1 == 'N':
        new = Add_Student(student,new_students)
        break

#call for the SEARCH Function
choice_2 =  input('\nDo you want to SEARCH for a Student(y/n) => ').upper()

if choice_2 == 'Y':
    name = input('\nEnter the name you want to SEARCH (full name with space between) => ')
    name = name.split(' ')
    Search_Student(new,name)

#call for the UPDATE Function
choice_3 =  input('\nDo you want to UPDATE a Student Record(y/n) => ').upper()

if choice_3 == 'Y':
    update_list=[]
    update_roll = input("\nEnter the Roll number => ")
    update_value = input("Enter the value to be updated(first name),(last name),(status),(cgpa),(semester) => ")
    new_value = input("Enter the new value => ")
    
    for i in update_roll,update_value,new_value:
        update_list.append(i)
    
    Update_Student(new,update_list)

#call for the DELETE Function
choice_4 =  input('\nDo you want to DELETE a Student Record(y/n) => ').upper()

if choice_4 == 'Y':
    delete_name = input("Enter the Name of the Student to DELETE (full name with space between) => ")
    delete_name = delete_name.split(' ')
    Delete_Student(new,delete_name)

#call for the GET Function
choice_5 =  input('\nDo you want to GET the Student Record(y/n) => ').upper()

if choice_5 == 'Y':
    Get_Students(student)

print('\n\n\t\t\tTHANK YOU ------> THE END')
```
